```python
import numpy as np
import pandas as pd
```

## Question 01

```python
data1 = {
    'Name': ['Alice', 'Bob', np.nan, 'David'],
    'Age': [25, np.nan, 22, 28],
    'City': ['Colombo', 'Kandy', 'Galle', np.nan],
    'Civil Status' : ['Married', np.nan, 'Single', np.nan]
}

df1 = pd.DataFrame(data1)

df1
```

```python
# a) Identify missing values
df1.isnull()
```

```python
# b) Fill missing values
df1.fillna({
    'Name': 'Unknown',
    'Age': df1['Age'].mean(),  # Mean age = (25 + 22 + 28)/3 = 25.0
    'City': 'Unknown'
}, inplace=True)
```

```python
df1
```

```python
# c) Drop rows with any missing values
df1.dropna()
```

## Question 02

```python
data2 = {
    'ID': [1, 2, 3, 3, 4, 5, 5],
    'Name': ['A', 'B', 'C', 'C', 'D', 'E', 'E']
}
df2 = pd.DataFrame(data2)
df2
```

```python
# a) Detect duplicate rows
print(f"These are :")
print(df2.duplicated())
duplicates = df2[df2.duplicated()]
duplicates
```

## Question 03

```python
data3 = {
    'ID': ['101', '102', '103'],
    'JoinDate': ['2022-01-10', '2022-02-15', '2022-03-01'],
    'Salary': ['50000', '60000', '70000']
}
df3 = pd.DataFrame(data3)
df3
```

```python
# Convert to appropriate data types
df3['ID'] = df3['ID'].astype(int)
df3['JoinDate'] = pd.to_datetime(df3['JoinDate'])
df3['Salary'] = df3['Salary'].astype(float)

df3
```

## Question 05

```python
data5 = {
    'Score': [55, 60, 62, 90, 95, 200, 58, 59, 210]
}
df5 = pd.DataFrame(data5)
df5
```

```python
# a) Detect values greater than a threshold (e.g., >100)
outliers = df5[df5['Score'] > 100]
outliers
```

```python
# b) Replace outliers with mean of non-outliers
mean_score = df5[df5['Score'] <= 100]['Score'].mean()  # = 68.428571
df5['Score'] = df5['Score'].apply(lambda x: mean_score if x > 100 else x)
df5
```
